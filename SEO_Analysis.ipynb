```python
import pandas as pd

# Load the dataset
df = pd.read_csv("big-sports.csv")

# Check basic info
print(df.info())
print(df.head())

# Check for missing or duplicate values
print(df.isnull().sum())
print(df.duplicated().sum())
```

```python
df['ctr'] = pd.to_numeric(df['ctr'], errors='coerce')
df['avg_position'] = pd.to_numeric(df['avg_position'], errors='coerce')
df['search_volume'] = pd.to_numeric(df['search_volume'], errors='coerce')
df
```

```python
df.dropna(inplace=True)

df.drop_duplicates(inplace=True)
df
```

# Analyze Current SEO Performance
# (a) Top Performing Keywords


```python
# Analyze Current SEO Performance
# (a) Top Performing Keywords
top_keywords = df.sort_values(by='clicks', ascending=False).head(10)
print(top_keywords[['keyword', 'clicks', 'ctr', 'avg_position', 'country']])
```

# (b) Best Performing Markets

```python
# (b) Best Performing Markets
country_perf = df.groupby('country')[['clicks','impressions']].sum().reset_index()
country_perf['ctr'] = country_perf['clicks'] / country_perf['impressions']
country_perf.sort_values(by='clicks', ascending=False)
```

# Identify SEO Growth Opportunities
# (a) High Volume, Low Ranking Keywords

```python
# Identify SEO Growth Opportunities
# (a) High Volume, Low Ranking Keywords
high_volume = df['search_volume'].median()
opportunities = df[(df['search_volume'] > high_volume) & (df['avg_position'] > 5)]
opportunities = opportunities.sort_values(by='search_volume', ascending=False)
print(opportunities[['keyword','search_volume','avg_position']].head(10))
```

# (b) Competitors Ranking Higher

```python
# (b) Competitors Ranking Higher
competitor_adv = df[df['competitor_position'] < df['avg_position']]
competitor_adv = competitor_adv.sort_values(by='competitor_position').head(10)
print(competitor_adv[['keyword','avg_position','competitor_position']])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Top keywords by clicks
plt.figure(figsize=(10,5))
sns.barplot(data=top_keywords, x='keyword', y='clicks', hue='country')
plt.xticks(rotation=45)
plt.title('Top 10 Keywords by Organic Clicks')
plt.show()

# Country performance
plt.figure(figsize=(6,4))
sns.barplot(data=country_perf, x='country', y='clicks')
plt.title('Organic Clicks by Country')
plt.show()

# CTR vs Position scatter
sns.scatterplot(data=df, x='avg_position', y='ctr', hue='country', size='search_volume')
plt.title('CTR vs Average Position')
plt.show()
```

# Estimate Potential Traffic Gain (Optional Model)

# Assume if you move from position 8 → 3, CTR improves from ~3% → ~10%.

```python
# Estimate Potential Traffic Gain (Optional Model)

# Assume if you move from position 8 → 3, CTR improves from ~3% → ~10%.

df['expected_ctr_if_rank3'] = 0.10
df['expected_clicks_if_rank3'] = df['impressions'] * df['expected_ctr_if_rank3']
df['click_gain'] = df['expected_clicks_if_rank3'] - df['clicks']

gain_table = df.sort_values(by='click_gain', ascending=False).head(10)
print(gain_table[['keyword','click_gain','avg_position']])
```
